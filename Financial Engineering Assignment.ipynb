Black-Scholes-Merton formula allows us to have a closed-form solutions for plain vanilla European call and put options written over nondividend paying securities. You will realize that using our risk neutral pricing tool we can extend those formulas for options on dividend paying securities. Under the risk-neutral measure, you will find that
\begin{equation*}
dS(t) = (r-a)S(t)dt + \sigma S(t)d\tilde{W}(t)
\end{equation*}
Again, the payoff of European call and put at maturity $T$ are given by:
\begin{gather*}
c(T, S(T))) = (S(T) - K)^{+} \\
p(T, S(T))) = (K - S(T))^{+}
\end{gather*}
Therefore, similar formulas to BSM formula can be obtained as you can see below
\begin{gather*}
c(t, x) = xe^{-a\tau}N(d_+(\tau,x)) - e^{-r\tau}KN(d_-(\tau,x))\\
p(t, x) = e^{-r\tau}KN(-d_-(\tau,x)) - xe^{-a\tau}N(-d_+(\tau,x))
\end{gather*}
where
\begin{equation*}
d_\pm(\tau,x) = \frac{1}{\sigma\sqrt{\tau}}\left[ln\frac{x}{K} + (r-a\pm\frac{1}{2}\sigma^2)\tau\right]
\end{equation*}
and $\tau = T - t, x$ can be replace by $S(t)$ and $N(\cdot)$ is the standard normal cdf.

### 2. (2 points) Write down the Greeks for these plain vanilla European call and put options, namely delta, gamma, vega, rho and theta.

The Greeks are written as given by Hull (2012, p 420). We consider $N(\cdot)$ to be the standard normal cumulative distribution function, and N'. the standard normal distribution function given by:
\begin{equation*}
N'(z) = \frac{e^{\frac{-z^2}{2}}}{\sqrt{2\pi}}
\end{equation*}
Then, the Greeks are as follows:

- Delta

\begin{gather*}
\text{Call Delta} = e^{-a\tau}N(d_1) \\
\text{Put Delta} = e^{-a\tau}(N(d_1)-1)
\end{gather*}

- Gamma

\begin{equation*}
\text{Gamma} = \frac{N'(d_1)e^{-a\tau}}{S_0\sigma\sqrt{\tau}}
\end{equation*}

The form of the Gamma is the same for both European calls and puts

- Vega

\begin{equation*}
\text{Vega} = S_0\sqrt{\tau}N'(d_1)e^{-a\tau}
\end{equation*}

The form of the Vega is the same for both European calls and puts

- Theta

\begin{gather*}
\text{Call Theta} = -\frac{S_0N'(d_1)\sigma e^{-a\tau}}{2\sqrt{\tau}} + aS_0N(d_1)e^{-a\tau} - rKe^{-r\tau}N(d_2) \\
\text{Put Theta} = -\frac{S_0N'(d_1)\sigma e^{-a\tau}}{2\sqrt{\tau}} - aS_0N(d_1)e^{-a\tau} + rKe^{-r\tau}N(-d_2)
\end{gather*}

- Rho

\begin{gather*}
\text{Call Rho} = K\tau e^{-r\tau}N(d_2) \\
\text{Put Rho} = -Ke^{-r\tau}N(-d_2)
\end{gather*}

```python
import numpy as np
import matplotlib.pyplot as plt
num_sim = 40  # num de simulaciones
n = 250       # Numero de pasos
mu = 0.05     # mu es la tasa risk-free en el periodo de tiempo establecido
sigma = 0.30  # volatilidad es la volatilidad impl√≠cita de la inversi√≥n  
S0 = 100
T = 0.5
K = 100
dt = T/n
np.random.seed(17)
z = np.random.randn(n,num_sim)
S = np.zeros((n,num_sim))
S[0,:] = S0

for i in range(1, n):
    S[i,:] = S[i - 1,:]*np.exp((mu - 0.5*sigma**2)*dt + sigma*np.sqrt(dt)*z[i - 1,:])

plt.plot(S)
plt.show(close=None, block=None)
```
